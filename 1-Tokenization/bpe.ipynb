## Neural Machine Translation of Rare Words with Subword Units


```python
sample_text = """
low low low low low
lower lower
newest newest newest newest newest newest
widest widest widest
"""
```

```python
import re

def word_leve_token(text):
    text = text.lower()
    tokens = re.findall(r"\w+|[^\w\s]", text)
    return tokens
```

```python
word_token = word_leve_token(sample_text)
word_token
```

```python
char_vocab = {int}
for token in word_token:
    char_seq = ' '.join(list(token))
    char_vocab[char_seq] += 1
```

```python
from collections import Counter

vocab = {
    'l o w': 5,
    'l o w e r': 2,
    'n e w e s t': 6,
    'w i d e s t': 3
}
```

```python
def get_pairs(vocab):
    pairs = Counter()
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[(symbols[i], symbols[i+1])] += freq
    return pairs
```

```python
get_pairs(vocab)
```

```python
def merge_vocab(vocab, pair):
    new_vocab = {}
    old = ' '.join(pair)
    new = ''.join(pair)
    for word, freq in vocab.items():
        new_word = word.replace(old, new)
        new_vocab[new_word] = freq
    return new_vocab
```

```python
merge_vocab(vocab, pairs)
```
